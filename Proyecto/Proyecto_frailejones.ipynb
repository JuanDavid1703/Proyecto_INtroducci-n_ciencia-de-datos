##Resumen de configuración de los datos:
Los datos consisten en 180 imágenes de hojas de frailejones de diferentes tamaños en 4 diferentes capas de colores, estas fueron obtenidas mediante fotografías tomadas en le páramo del Sumapaz, cada fotografía tenpia en promedio de 3 a 4 hojas de frailejón.  

De estos datos se desea realizar una apropiada clasificación de especies de frailejones mediante la caracterización y clasificación de hojas de frailejón usando morfología foliar. 

Como primicia se quizo reconocer los tipos de datos que se tenían, para ello se debió separar las hojas del fondo de cada fografía, para este proceso se usó "Adobe photshop express", programa que fue escogido después de intentar con varios, una de las más fuertes complicaciones en la separación de las hojas del fondo se debía a la calidad de la iluminación de cada imagen, como no se tenía buena iluminación, no se pudieron separar todas las hojas de todas las imágenes, se lograron obtener 314 hojas de frailejón separadas y en su mayoría bien contrastadas con su fondo.

Después de esto la idea era obtener los contornos de las hojas para poder caracterizar cada hoja usando la tranformada de Fourier rápida, esto también tuvo dificultades dado el ruido que había quedado en la mayoría de imágenes y los colores tenues que se encontraban en algunas hojas, por esto se tuvo que proceder en realizar una limpieza exhaustiva y además arreglar las tonalidades del color de cada imagen.


Depués de haber obtenido con éxito los contornos se procedió a hallar características extras de cada hoja, las características escojidas fueron:

* El área
* El perímetro
* Las medidas de los ejes axiales, el mayor y el menor.
* La eccentricidad
* El centroide   


Las anteiores caracteristicas fueron escogidas dado que cada hoja se puede encerrar en una elipse que está caracterizada por ciertos parámetros, esto nos permitiría caracterizar mejor cada hoja de frailejón, además como otra medida, al usar los contornos se pueden identificar las curvas que representan cada componenete del conjunto de puntos como se ve en la imagen siguiente:

**Momentos Hu**: 

Los momentos de una imagen son un conjunto de caracterizadores de datos en segunda dimensión que usualamente  funcionan para identificar numéricamente una imagen, los momentos de una función bivariada estan definidos por 
$$\mu_{pq}=\int_{-\infty}^{\infty}\int_{-\infty}^{\infty}x^py^q f(x,y)dxdy$$


Los momentos que se caracterizan por ser invariantes bajo translaciones y escalación uniforme pero so son afectados por la rotaciones, una clásica solución de este problema es una combinación lineal  de estos momentos, estas combinaciones son llamadas momentos de Hu mostrados a continuación:

<img title="Páramo" src="https://github.com/JuanDavid1703/Proyecto_INtroducci-n_ciencia-de-datos/raw/main/Proyecto/Imagenes/Hu.png" width="500" height="500">

A partir de las anteriores diagramas se puede observar que existe una considerable variación en los momentos de Hu, eso quiere decir que los momentos caracterizan de una buena manera ls imágenes.

Armonicos de Fourier:

Dada una función, la transfomada de Fourier permite describir esa función en terminos de una serie de funciones trigonométricas en donde su representación esta dada por el espacio de espectro, este espacio lo conforman los armónicos.

En el análisis de imágenes se usa la transfomada rápida de Fourier, un mecanismo que permite analizar y caracterizar el contorno o los bordes de una imagen binarizada

$$X_k=\sum_{n=0}^{N-1}{x_n e^{\frac{-2\pi i}{N}kn}}$$

con $k=1,2,3 \ldots N-1$

Los armónicos permiten identificar puntas y bordes suaves, cuando los armónicos son de valores de alta variación, eso podría representar puntas o esquinas en las imágenes pero si son de baja variabilidad podria representar bordes suaves con curvas suaves.

en la siguiente imagen se puede apreciar 4 diferentes representaciones de los armónicos de los bordes 4 distintas hojas tomadas al azar y se puede observar que en dos de las cuatro imágenes hay mayor variación entre los valores que las otras dos:


<img title="Páramo" src="https://github.com/JuanDavid1703/Proyecto_INtroducci-n_ciencia-de-datos/raw/main/Proyecto/Imagenes/Armonicos.png" width="900" height="600">


Para poder escoger unas adecuadas caracterpistica de deberían tomar los armónicos de mayor variación, se debe tener en cuenta la dimensionalidad de las características por ende se deben tomar menos armónicos para el espacio de características.
